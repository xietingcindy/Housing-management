```python
# 导入数据库数据
import numpy as np
import pandas as pd
import pymysql
conn = pymysql.connect(
    host='localhost',
    user='root',
    password='123456789',
    database='house_datas',
    charset='utf8'
)
df = pd.read_sql('select * from house1',conn)
df
```

```python
# 拷贝一份数据，不对原始数据进行修改
df1 = df.copy()
df1
```

```python
# 删除重复数据,保存第一次出现的一行
df1.drop_duplicates(subset=['city','dist','flag','title','address','price','size','style','floor','href'],keep='first',inplace=True)
df1
```

```python
# 填充字段，计算 reality=总价 / 总面积
df1['reality'] = df1['price'] / df1['size']
df1
```

```python
#以title关键字排序，方便查看去重是否成功
df1.sort_values("title",inplace=True)
df1
```

```python
# 空值：在pandas中的空值是""  缺失值：在dataframe中为nan或者naT（缺失时间），在series中为none或者nan即可
# 判断含有缺失值的列，朝向和楼层可以缺失不作处理
print(df1.isnull().any())
```

```python
# 删除dist为''的行
df1=df1[~df1['dist'].isin([''])]
df1
```

```python
#普通房屋面积不会超过1000平方米，所以先删除面积大于1000的房屋
condition =  (df1['size'] <1000)
df1 = df1[condition]
df1
```

```python
# 研究房屋住房面积与租金的关系
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
import seaborn as sns
from IPython.display import display
plt.style.use("fivethirtyeight")
sns.set_style({'font.sans-serif':['simhei','Arial']})


#绘制散点图
df1.plot.scatter(x='size', y='price')
#设置坐标轴范围
plt.xlim((0, 900))
plt.ylim((0,110000))
```

```python
# 数据库中观察到有面积为0的异常房源，在此删除size关键字段为0的异常值
df1=df1[~df1['size'].isin([0])]
df1
```

```python
# 观察散点图有一个价格低于20000面积大于800的异常值，将其删除
df_clear = df1.drop(df1[(df1['price']<20000) & (df1['size']>800)].index)
df_clear
```

```python
# 分析数据特征，选择方法进行异常排除
df_clear['reality'].plot(kind='hist')  #查看分布柱状图
```

```python
# 判断数据是否服从正态分布，服从则采用二倍标准差法，不服从则使用箱型图法
#标准差法，就是用以样本均值+样本标准差为基准，如果样本离平均值相差2个标准差以上的就是异常值
#箱形图法：以上下四分位作为参考， x > Q3+nIQR 或者 x < Q1 - nIQR 简单地理解，就是如果样本值不在上下四分位+标准差范围内，就是异常值
#两种异常值判定是，如数据近似服从正态分布是，优先选择n个标准差法，因为数据的分布相对比较对称：否则优先选择箱线图法，因为分位数并不会受极端值的影响。

#计算avg,std(平均值、标准差)
avg = df_clear['reality'].mean()
avg
```

```python
std = df_clear['reality'].std()
std
```

```python
# 判断数据是否服从正态分布，KS检验
# .kstest方法：KS检验，参数分别是：待检验的数据，检验方法（这里设置成norm正态分布），均值与标准差
# 结果返回两个值：statistic → D值，pvalue → P值
# p值大于0.05，为正态分布
from scipy import stats
stats.kstest(df_clear['reality'], 'norm', (avg, std))
```

```python
# 数据分布图形检验
import matplotlib.pyplot as plt
plt.style.use('ggplot')
df_clear.reality.plot(kind='hist',bins = 30,density = True)
df_clear.reality.plot(kind='kde')
plt.show()
```

```python
# 数据整体偏右，且不服从正态分布，故采用四分位法判断异常值
# 四分位数展布法，reality字段从小到大排序
df_clear.sort_values("reality",inplace=True)
df_clear
```

```python
Q1 = df_clear.reality.quantile(q = 0.25)
Q1
```

```python
Q3 = df_clear.reality.quantile(q = 0.75)
Q3
```

```python
IQR = Q3 -Q1
IQR
```

```python
low = Q1-1.5*IQR   #范围最小
low
```

```python
high = Q3 + 1.5*IQR    #范围最大
high
```

```python
any(df_clear['reality']<low) #检测有无小于最小边界的异常值
```

```python
any(df_clear['reality']>high) #检测有无大于最大边界的异常值
```

```python
#异常值偏大，删除偏大的结果，得出最终数据
result = df_clear.loc[~(df_clear['reality'] > high)]
result
```

```python
#重新查看散点图
result.plot.scatter(x='size', y='price')
#设置坐标轴范围
plt.xlim((0, 900))
plt.ylim((0,110000))
```

```python
# 将清洗结果存入数据库
from sqlalchemy import create_engine

#转换类型，改变默认数据库表字段输入类型 
from sqlalchemy.types import VARCHAR, Float, Integer 
def mapping_df_types(df):
    dtypedict = {}
    for i, j in zip(df.columns, df.dtypes):
        if "object" in str(j):
            dtypedict.update({i: VARCHAR(length=255)})
        if "float" in str(j):
            dtypedict.update({i: Float(precision=2, asdecimal=True)})
        if "int" in str(j):
            dtypedict.update({i: Integer()})
    return dtypedict
dtypedict = mapping_df_types(result)

#新建house表插入指定字段数据
engine = create_engine('mysql+pymysql://root:123456789@localhost:3306/house_datas')
con = engine.connect()
result.to_sql(name='house', 
          con=con,
          if_exists='replace', 
          index=False,
          dtype=dtypedict
         )

print('Read from and write to Mysql table successfully!')
```
