```python
#对house表进行数据统计，得出租房全览所需的watch0,watch1,watch2表
import numpy as np
import pandas as pd
import pymysql
conn = pymysql.connect(
    host='localhost',
    user='root',
    password='123456789',
    database='house_datas',
    charset='utf8'
)
df = pd.read_sql('select * from house',conn)
df
```

```python
# 拷贝一份数据，不对原始数据进行修改
df1 = df.copy()
df1
```

```python
#筛选出独栋数据，独栋无地区信息
df2 = df1.loc[df1['flag'] == '独栋']
df2
```

```python
# 有地区信息
df3 = df1.loc[df1['flag'] != '独栋']
df3
```

```python
#筛选出北京的数据
city0 = df3.loc[df3['city'] == '0']
city0
```

```python
# count()统计数量，
# sort_values(ascending=True)升序排列
# to_frame()将series转换为dataframe
# count字段
house_count = city0.groupby('dist')['reality'].count().sort_values(ascending=True).to_frame()
# 修改列名称
house_count = house_count.reset_index().rename({'dist':'dist','reality':'count'},axis=1).set_index('dist')
#按地区排序，方便合并
house_count.sort_values("dist",inplace=True)
house_count
```

```python
#每平米单价字段，按地区排序
mean_house = city0.groupby('dist')['reality'].mean().to_frame().reset_index().rename({'dist':'dist','reality':'perprice'},axis=1).set_index('dist')
mean_house.sort_values("dist",inplace=True)
mean_house
```

```python
#合并两表的值
house_count['perprice'] = mean_house['perprice']
#将dist作为列字段
house_count['dist'] = house_count.index
```

```python
#添加city字段
house_count['city'] = '0'
house_count #得到北京地区信息
```

```python
#转换类型，改变默认数据库表字段输入类型 
from sqlalchemy.types import VARCHAR, Float, Integer 
from sqlalchemy import create_engine
def mapping_df_types(df):
    dtypedict = {}
    for i, j in zip(df.columns, df.dtypes):
        if "object" in str(j):
            dtypedict.update({i: VARCHAR(length=255)})
        if "float" in str(j):
            dtypedict.update({i: Float(precision=2, asdecimal=True)})
        if "int" in str(j):
            dtypedict.update({i: Integer()})
    return dtypedict
dtypedict = mapping_df_types(dist_info)

#将结果插入watch0表
engine = create_engine('mysql+pymysql://root:123456789@localhost:3306/house_datas')
con = engine.connect()
dist_info.to_sql(name='watch0', 
          con=con,
          if_exists='replace', 
          index=False,
          dtype=dtypedict
         )

print('Read from and write to Mysql table successfully!')
```
